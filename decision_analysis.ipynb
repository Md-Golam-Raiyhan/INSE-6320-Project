```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# ==========================================
# Output folder
# ==========================================
output_dir = Path("decision_tree_outputs")
output_dir.mkdir(exist_ok=True)

# ==========================================
# 1. Define probabilities and payoffs
# ==========================================
probabilities = {
    "Low Risk": 0.30,
    "Medium Risk": 0.50,
    "High Risk": 0.20
}

payoff_df = pd.DataFrame({
    "Strategy": [
        "Base Strategy",
        "Moderate Mitigation",
        "Aggressive Mitigation"
    ],
    "Low Risk": [118, 121, 125],
    "Medium Risk": [129, 127, 128],
    "High Risk": [142, 135, 131]
})

# ==========================================
# 2. Create branch-level decision tree table
# ==========================================
rows = []

for _, row in payoff_df.iterrows():
    strategy = row["Strategy"]
    for scenario in ["Low Risk", "Medium Risk", "High Risk"]:
        rows.append({
            "Strategy": strategy,
            "Scenario": scenario,
            "Probability": probabilities[scenario],
            "Payoff_Cost_MUSD": row[scenario]
        })

decision_tree_table = pd.DataFrame(rows)
decision_tree_table.to_csv(output_dir / "table_decision_tree_branches.csv", index=False)

print("\nDecision Tree Branch Table")
print(decision_tree_table)

# ==========================================
# 3. Plot simple decision tree
# ==========================================
fig, ax = plt.subplots(figsize=(12, 7))
ax.axis("off")

# positions
decision_x, decision_y = 0.1, 0.5

strategy_positions = {
    "Base Strategy": (0.35, 0.75),
    "Moderate Mitigation": (0.35, 0.50),
    "Aggressive Mitigation": (0.35, 0.25)
}

scenario_offsets = {
    "Low Risk": 0.12,
    "Medium Risk": 0.00,
    "High Risk": -0.12
}

# draw main decision node
ax.text(decision_x, decision_y, "Decision", ha="center", va="center",
        bbox=dict(boxstyle="round,pad=0.3", fc="lightgray", ec="black"))

# draw strategy nodes and branches
for strategy, (sx, sy) in strategy_positions.items():
    # branch from decision node to strategy
    ax.plot([decision_x, sx], [decision_y, sy], color="black")
    ax.text((decision_x + sx)/2 - 0.02, (decision_y + sy)/2 + 0.02, strategy,
            fontsize=9, ha="center")

    # strategy node
    ax.text(sx, sy, "Chance", ha="center", va="center",
            bbox=dict(boxstyle="circle,pad=0.25", fc="white", ec="black"))

    # get payoffs
    row = payoff_df[payoff_df["Strategy"] == strategy].iloc[0]

    # draw scenario branches
    for scenario, offset in scenario_offsets.items():
        ex, ey = 0.75, sy + offset
        payoff = row[scenario]
        prob = probabilities[scenario]

        ax.plot([sx, ex], [sy, ey], color="black")
        ax.text((sx + ex)/2, (sy + ey)/2 + 0.02,
                f"{scenario}\np={prob}",
                fontsize=8, ha="center")

        ax.text(ex, ey, f"Cost={payoff}", ha="center", va="center",
                bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="black"))

ax.set_title("Decision Tree for Metro Rail Mitigation Strategies", pad=20)
plt.tight_layout()
plt.savefig(output_dir / "figure_decision_tree.png", dpi=300, bbox_inches="tight")
plt.close()

print("\nFiles saved in:", output_dir)
```

```python
import pandas as pd
from pathlib import Path

# ==========================================
# Output folder
# ==========================================
output_dir = Path("expected_value_outputs")
output_dir.mkdir(exist_ok=True)

# ==========================================
# 1. Define payoff table
# ==========================================
payoff_df = pd.DataFrame({
    "Strategy": [
        "Base Strategy",
        "Moderate Mitigation",
        "Aggressive Mitigation"
    ],
    "Low Risk": [118, 121, 125],
    "Medium Risk": [129, 127, 128],
    "High Risk": [142, 135, 131]
})

# save payoff table
payoff_df.to_csv(output_dir / "table_payoff_for_ev.csv", index=False)

# ==========================================
# 2. Define scenario probabilities
# ==========================================
probabilities = {
    "Low Risk": 0.30,
    "Medium Risk": 0.50,
    "High Risk": 0.20
}

prob_df = pd.DataFrame({
    "Scenario": list(probabilities.keys()),
    "Probability": list(probabilities.values())
})
prob_df.to_csv(output_dir / "table_scenario_probabilities.csv", index=False)

# ==========================================
# 3. Calculate expected values
# ==========================================
payoff_df["Expected Cost (MUSD)"] = (
    payoff_df["Low Risk"] * probabilities["Low Risk"] +
    payoff_df["Medium Risk"] * probabilities["Medium Risk"] +
    payoff_df["High Risk"] * probabilities["High Risk"]
)

# rank strategies (lower expected cost is better)
payoff_df["Rank"] = payoff_df["Expected Cost (MUSD)"].rank(method="dense", ascending=True).astype(int)

# sort by expected cost
ev_df = payoff_df[["Strategy", "Expected Cost (MUSD)", "Rank"]].sort_values(
    "Expected Cost (MUSD)", ascending=True
).reset_index(drop=True)

# round values
ev_df["Expected Cost (MUSD)"] = ev_df["Expected Cost (MUSD)"].round(3)

# save results
ev_df.to_csv(output_dir / "table_expected_value_results.csv", index=False)

print("\nExpected Value Results")
print(ev_df)

# ==========================================
# 4. Optional detailed calculation table
# ==========================================
calc_df = payoff_df.copy()
calc_df["EV Calculation"] = (
    "(" + str(probabilities["Low Risk"]) + "×" + calc_df["Low Risk"].astype(str) + ") + "
    "(" + str(probabilities["Medium Risk"]) + "×" + calc_df["Medium Risk"].astype(str) + ") + "
    "(" + str(probabilities["High Risk"]) + "×" + calc_df["High Risk"].astype(str) + ")"
)
calc_df.to_csv(output_dir / "table_expected_value_calculations.csv", index=False)

print("\nFiles saved in:", output_dir)
```
