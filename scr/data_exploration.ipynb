```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# ==========================================
# Load dataset
# ==========================================
file_path = "metro_rail_project_dataset.csv"
df = pd.read_csv(file_path)

# output folder
output_dir = Path("better_correlation_outputs")
output_dir.mkdir(exist_ok=True)

# ==========================================
# Select numerical columns
# ==========================================
numeric_cols = [
    "segment_length_km",
    "baseline_cost_musd",
    "planned_duration_months",
    "material_price_index",
    "labor_availability_index",
    "weather_delay_days",
    "permit_delay_days",
    "equipment_downtime_hours",
    "inflation_rate_annual",
    "contractor_experience_score",
    "ground_risk_score",
    "change_orders",
    "safety_incidents",
    "actual_duration_months",
    "actual_cost_musd",
    "cost_overrun_musd",
    "cost_overrun_pct"
]

data = df[numeric_cols]

# ==========================================
# 1. Full correlation matrix
# ==========================================
corr_matrix = data.corr()

# save full matrix
corr_matrix.to_csv(output_dir / "full_correlation_matrix.csv")

# ==========================================
# 2. Filter weak correlations
#    keep only strong ones
# ==========================================
threshold = 0.30   # you can change this to 0.25 or 0.40
filtered_corr = corr_matrix.copy()

for i in range(filtered_corr.shape[0]):
    for j in range(filtered_corr.shape[1]):
        if abs(filtered_corr.iloc[i, j]) < threshold and i != j:
            filtered_corr.iloc[i, j] = np.nan

# save filtered matrix
filtered_corr.to_csv(output_dir / "filtered_correlation_matrix.csv")

# ==========================================
# 3. Plot filtered correlation heatmap
# ==========================================
plt.figure(figsize=(11, 8))
im = plt.imshow(filtered_corr, cmap="coolwarm", vmin=-1, vmax=1, aspect="auto")

plt.colorbar(im, label="Correlation")

plt.xticks(range(len(filtered_corr.columns)), filtered_corr.columns, rotation=90)
plt.yticks(range(len(filtered_corr.index)), filtered_corr.index)

plt.title(f"Filtered Correlation Matrix (|r| >= {threshold})")

# annotate only strong values
for i in range(filtered_corr.shape[0]):
    for j in range(filtered_corr.shape[1]):
        val = filtered_corr.iloc[i, j]
        if pd.notna(val) and i != j:
            plt.text(j, i, f"{val:.2f}", ha="center", va="center", fontsize=8)

plt.tight_layout()
plt.savefig(output_dir / "filtered_correlation_heatmap.png", dpi=300)
plt.close()

# ==========================================
# 4. Top correlations with cost overrun
# ==========================================
target = "cost_overrun_musd"

corr_with_target = corr_matrix[target].drop(target)
corr_with_target = corr_with_target.reindex(corr_with_target.abs().sort_values(ascending=False).index)

# save table
corr_with_target_df = corr_with_target.reset_index()
corr_with_target_df.columns = ["Variable", "Correlation_with_Cost_Overrun"]
corr_with_target_df.to_csv(output_dir / "top_correlations_with_cost_overrun.csv", index=False)

print("\nTop correlations with cost_overrun_musd:")
print(corr_with_target_df)

# ==========================================
# 5. Plot top correlations with cost overrun
# ==========================================
plt.figure(figsize=(9, 6))
corr_with_target.sort_values().plot(kind="barh")
plt.title("Top Correlations with Cost Overrun")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Variables")
plt.tight_layout()
plt.savefig(output_dir / "top_correlations_with_cost_overrun.png", dpi=300)
plt.close()

print("\nOutputs saved in folder:", output_dir)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# ==========================================
# Load dataset
# ==========================================
file_path = "metro_rail_project_dataset.csv"
df = pd.read_csv(file_path)

# output folder
output_dir = Path("reduced_correlation_outputs")
output_dir.mkdir(exist_ok=True)

# ==========================================
# Select numerical columns
# ==========================================
numeric_cols = [
    "segment_length_km",
    "baseline_cost_musd",
    "planned_duration_months",
    "material_price_index",
    "labor_availability_index",
    "weather_delay_days",
    "permit_delay_days",
    "equipment_downtime_hours",
    "inflation_rate_annual",
    "contractor_experience_score",
    "ground_risk_score",
    "change_orders",
    "safety_incidents",
    "actual_duration_months",
    "actual_cost_musd",
    "cost_overrun_musd",
    "cost_overrun_pct"
]

data = df[numeric_cols]

# ==========================================
# Correlation with target
# ==========================================
target = "cost_overrun_musd"
corr_with_target = data.corr()[target].drop(target)

# choose threshold
threshold = 0.15   # change if needed

# keep only important variables
important_vars = corr_with_target[abs(corr_with_target) >= threshold].index.tolist()

# add target itself
important_vars = important_vars + [target]

print("Variables kept in reduced correlation matrix:")
print(important_vars)

# save list
pd.DataFrame({"Selected_Variables": important_vars}).to_csv(
    output_dir / "selected_variables_for_reduced_correlation.csv",
    index=False
)

# ==========================================
# Reduced correlation matrix
# ==========================================
reduced_corr = data[important_vars].corr()
reduced_corr.to_csv(output_dir / "reduced_correlation_matrix.csv")

# ==========================================
# Plot reduced heatmap
# ==========================================
plt.figure(figsize=(8, 6))
im = plt.imshow(reduced_corr, cmap="coolwarm", vmin=-1, vmax=1, aspect="auto")

plt.colorbar(im, label="Correlation")

plt.xticks(range(len(reduced_corr.columns)), reduced_corr.columns, rotation=90)
plt.yticks(range(len(reduced_corr.index)), reduced_corr.index)

plt.title(f"Reduced Correlation Matrix (|r with cost overrun| >= {threshold})")

# annotate values
for i in range(reduced_corr.shape[0]):
    for j in range(reduced_corr.shape[1]):
        plt.text(j, i, f"{reduced_corr.iloc[i, j]:.2f}",
                 ha="center", va="center", fontsize=8)

plt.tight_layout()
plt.savefig(output_dir / "reduced_correlation_heatmap.png", dpi=300)
plt.close()

# ==========================================
# Optional: Bar plot of correlations with target
# ==========================================
selected_corr = corr_with_target[abs(corr_with_target) >= threshold]
selected_corr = selected_corr.reindex(selected_corr.abs().sort_values(ascending=False).index)

plt.figure(figsize=(8, 5))
selected_corr.sort_values().plot(kind="barh")
plt.title(f"Selected Correlations with {target}")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Variables")
plt.tight_layout()
plt.savefig(output_dir / "selected_correlations_with_target.png", dpi=300)
plt.close()

print("\nOutputs saved in:", output_dir)
```
